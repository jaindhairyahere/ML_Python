```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.datasets import load_boston
from sklearn.metrics import explained_variance_score,mean_squared_error,r2_score
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
```

```python
housing_data = load_boston()
```

```python
X = pd.DataFrame(shuffle(load_boston().data, random_state=7))
y = pd.DataFrame(shuffle(load_boston().target, random_state=7))
```

```python
X=X.T
X.index = load_boston().feature_names
```

```python
X = X.T
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=40)
```

```python
dt_regressor = DecisionTreeRegressor(max_depth=4)
dt_regressor.fit(X_train,y_train)
```

```python
adaBoost_regressor = AdaBoostRegressor(DecisionTreeRegressor(max_depth=4),n_estimators=400,random_state=7)
adaBoost_regressor.fit(X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor(n_estimators=1000,max_depth=50,min_samples_split=1.0)
rf_regressor.fit(X_train,y_train)
```

```python
y_predicted_dt = dt_regressor.predict(X_test)
y_predicted_ada = adaBoost_regressor.predict(X_test)
y_predicted_rf = rf_regressor.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_predicted_dt)
evs = explained_variance_score(y_test, y_predicted_dt)
print ("\n#### Decision Tree performance ####")
print ("Mean squared error =", round(mse, 2))
print ("Explained variance score =", round(evs, 2))
```

```python
mse = mean_squared_error(y_test, y_predicted_ada)
evs = explained_variance_score(y_test, y_predicted_ada)
print ("\n#### AdaBoost performance ####")
print ("Mean squared error =", round(mse, 2))
print ("Explained variance score =", round(evs, 2))
```

```python
mse = mean_squared_error(y_test, y_predicted_rf)
evs = explained_variance_score(y_test, y_predicted_rf)
print ("\n#### Decision Tree performance ####")
print ("Mean squared error =", round(mse, 2))
print ("Explained variance score =", round(evs, 2))
```

```python
def plot_feature_importances(feature_importance,title,feature_names):
    # NOrmalize importace Values
    feature_importance = 100*((feature_importance)/max(feature_importance))
    # Sort the index values and flip them so that they are arranged in decreasing order of importance
    index_sorted = np.flipud(np.argsort(feature_importance))
    
    # Center the location of the labels on the X-axis (for displaypurposes only)
    pos = np.arange(index_sorted.shape[0]) + 0.5
    
    #plot Bar Graph
    plt.figure()
    plt.bar(pos,feature_importance[index_sorted], align='center')
    plt.xticks(pos,feature_names[index_sorted])
    plt.ylabel('Relative Importance')
    plt.title(title)
    plt.show()
```

```python
plot_feature_importances(dt_regressor.feature_importances_,'Decision Tree regressor', housing_data.feature_names)
plot_feature_importances(adaBoost_regressor.feature_importances_,'AdaBoost regressor', housing_data.feature_names)
```
